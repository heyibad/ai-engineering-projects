```python
display(df.columns)
# ['age', 'sex', 'bmi', 'children', 'smoker', 'region', 'charges']

numeric_cols = [  'age',    'bmi',    'children',    'charges']
categorical_cols = [  'sex',    'smoker',    'region']
```

```python
for n_col in numeric_cols:
    plt.figure(figsize=(6, 4)) # 600 x 400 pixels
    sns.histplot(df[n_col], bins=20, kde=True, color='blue', edgecolor='black', alpha=0.7)
```

```python
for n_col in categorical_cols:
    plt.figure(figsize=(6, 4)) # 600 x 400 pixels
    sns.countplot(data=df, x=n_col, palette='viridis', edgecolor='black', alpha=0.7)
    plt.title(f'Distribution of {n_col}')
    plt.xlabel(n_col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3) # alpha for grid transparency
    plt.show()
```

```python
sns.countplot(data=df, x='children', palette='rainbow', edgecolor='black', alpha=0.7)
```

```python
for n_col in numeric_cols:
    plt.figure(figsize=(6, 4)) # 600 x 400 pixels
    sns.boxplot(data=df, x=n_col, palette='Set2')
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linecolor='black')
plt.title('Correlation Heatmap')
```
