```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df["species"].unique()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['species'])
y = df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

model_svm = SVC(C=30, gamma="auto",kernel='rbf')
```

```python
model_svm.fit(X_train, y_train)

model_svm.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

classifer = GridSearchCV((model_svm),{
    "C":[1,2,10,20,30],
    "kernel":['rbf','linear']
}, cv=5, return_train_score=False)
```

```python
classifer.fit(X, y)
```

```python
res =pd.DataFrame(classifer.cv_results_)
```

```python
res
```

```python
res[["param_C","mean_test_score","param_kernel"]]
```

```python
from sklearn.model_selection import RandomizedSearchCV

classifer = RandomizedSearchCV((model_svm),{
    "C":[1,2,10,20,4,6,30],
    "kernel":['rbf','linear']
}, n_iter=10, cv=5, return_train_score=False)
classifer.fit(X, y)
```

```python
pd.DataFrame(classifer.cv_results_)[["param_C","mean_test_score","param_kernel"]]
```
