```python
def extract_ep(title:str)-> str:
  check = False;
  final_title=""
  for i in title:
    if i == "(":
      check =True
      continue;
    if check:
       final_title+=i;

       if i == " ":
        check = False;


  return int(final_title.strip())
```

```python
extract_ep('Fullmetal Alchemist: BrotherhoodTV (64 eps)Apr 2009 - Jul 20103,218,472 membersManga StoreVolume 14.58Preview')
```

```python
df["Episodes"] = df["Title"].apply(extract_ep)
```

```python
df
```

```python
def extract_timestamps(title:str)-> str:
  check = False;
  final_title=""
  for i in title:
    if i == ")":
      check =True
      continue;
    if check:
       final_title+=i;

  return final_title[0:19].strip()
```

```python
extract_timestamps('Fullmetal Alchemist: BrotherhoodTV (64 eps)Apr 2009 - Jul 20103,218,472 membersManga StoreVolume 14.58Preview')
```

```python
df["Timestamp"] = df["Title"].apply(extract_timestamps)
```

```python
def starting_date(title:str)->str:
  return title.split("-")[0].strip()

def end_date(t:str)->str:
  return t.split("-")[1]
```

```python
df["Start date"] = df["Timestamp"].apply(starting_date)
df["End date"] = df["Timestamp"].apply(end_date)
```

```python
from dateutil import parser
from dateutil.relativedelta import relativedelta

def duration_in_months(timestamp):
    start_str, end_str = timestamp.split(" - ")
    start = parser.parse(start_str)
    # print(start) # 2009-04-19 00:00:00
    end = parser.parse(end_str)
    # print(end) # 2010-07-19 00:00:00
    diff = relativedelta(end, start)
    # print(diff) # relativedelta(years=+1, months=+3)
    return diff.years * 12 + diff.months
```

```python
duration_in_months('Apr 2009 - Jul 2010')
```

```python
df["Duration in months"] = df["Timestamp"].apply(duration_in_months)
```
